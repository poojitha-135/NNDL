```python
#1a. Backpropagation Without Using Packages
import numpy as np
# Network configuration
inputs = 3
hidden_neurons = 3
outputs = 1
lr = 0.1
epochs = 1200
# Initialize weights and biases
np.random.seed(2)
W_hidden = np.random.rand(inputs, hidden_neurons)
b_hidden = np.random.rand(hidden_neurons)
W_out = np.random.rand(hidden_neurons, outputs)
b_out = np.random.rand(outputs)
# Input and target
X = np.array([[2, 4, 6]])
y = np.array([[0.8]])
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
# Training loop
for _ in range(epochs):
    # Forward pass
    h_input = np.dot(X, W_hidden) + b_hidden
    h_output = sigmoid(h_input)
    o_input = np.dot(h_output, W_out) + b_out
    prediction = sigmoid(o_input)
    # Error calculation
    error = y - prediction
    # Backward pass
    d_output = error * prediction * (1 - prediction)
    d_hidden = np.dot(d_output, W_out.T) * h_output * (1 - h_output)
    # Update weights and biases
    W_out += lr * np.dot(h_output.T, d_output)
    b_out += lr * d_output.flatten()
    W_hidden += lr * np.dot(X.T, d_hidden)
    b_hidden += lr * d_hidden.flatten()
print("Final Output:", prediction)
print("Final Error:", error)
```

```python
#1b. Backpropagation Using TensorFlow
import tensorflow as tf
import numpy as np
# Training data
X = np.array([[2, 4, 6]], dtype=float)
y = np.array([[0.8]], dtype=float)
# Model definition
model = tf.keras.Sequential([
    tf.keras.layers.Dense(3, activation='sigmoid', input_shape=(3,)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
# Compile model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mean_squared_error'
)
# Train model
model.fit(X, y, epochs=1200, verbose=0)
# Prediction
output = model.predict(X)
print("Predicted Output:", output)
```
