```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
# Generate dataset
data = {
    'amount': np.random.randint(500, 15000, 1200),
    'hour': np.random.randint(0, 24, 1200),
    'region': np.random.randint(1, 6, 1200),
    'label': np.random.randint(0, 2, 1200)
}
df = pd.DataFrame(data)
X = df[['amount', 'hour', 'region']]
y = df['label']
# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=1
)
# Scale features
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
# Build feedforward neural network
model = tf.keras.Sequential([
    tf.keras.layers.Dense(12, activation='relu', input_shape=(3,)),
    tf.keras.layers.Dense(6, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
# Compile model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
# Train model
model.fit(X_train, y_train, epochs=12, verbose=0)
# User input
amt = float(input("Enter amount: "))
hr = float(input("Enter time: "))
reg = float(input("Enter region: "))
user = scaler.transform([[amt, hr, reg]])
#Display results
result = model.predict(user)[0][0]
print("Fraud Score:", round(result, 3))
print("Status:", "Fraud" if result > 0.5 else "Legitimate")
```
